```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import json
import pandas as pd
```

```python
with open('nba_teams.json') as f:
    data = json.load(f)
team_key = list(data.keys())
team_url = list(data.values())
f.close()
```

```python
all_players = {}
for i in range(len(team_url)):
    bball_data = uReq(team_url[i])
    html = bball_data.read()
    bball_data.close()
   
    team_soup = bs(html,'html.parser')
    tables = team_soup.findAll('table')
    
    print(team_key[i])
    
    team = {}
    for y in range(1,len(tables[0].findAll('tr'))):
        
        player_name = tables[0].findAll('tr')[y].a.text
        player_url = 'https://basketball-refernce.com' + tables[0].findAll('tr')[y].a.get('href')
        team[','.join([team_key[i],player_name])] = player_url
    
    all_players.update(team)
        
```

```python
all_players
```

```python
j = json.dumps(all_players, indent=4)
with open('all_players.json', 'w') as f:
    f.write(j)
    f.close()
```
